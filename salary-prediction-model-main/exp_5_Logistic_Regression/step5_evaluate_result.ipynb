```python
import numpy
import pandas
import joblib
import json
import seaborn
import sklearn

%matplotlib inline
```

# Instructions

- Read **the test data** from the CSV file and properly set the index
- Load the trained model

```python
data_test= pandas.read_csv('./data/features.test.csv').set_index('id')
data_test
```

```python
model = joblib.load('./model/logistic_regression_model.joblib')
model
```


# Instructions

- Calculate and add the `prediction` column to the dataframe of test data
- Compare the labels with the predictions and put them into 4 `result_type` categories; namely,
  - true_negative
  - false_negative
  - true_positive
  - false_positive
- Add the `result_type` column to the dataframe of test data
- Write the dataframe to `./results/predictions.test.csv`

```python
data_test['prediction'] = model.predict(
     X= data_test.drop(columns=['label']),
)
data_test
```

```python
result_types =[
    {
        'prediction' : 0,
        'label' : 0,
        'result_type': 'true_negative'
    },
    {
        'prediction' : 1,
        'label' : 0,
        'result_type': 'false_positive'
    },
    {
        'prediction' : 0,
        'label' : 1,
        'result_type': 'false_negative'
    },
    {
        'prediction' : 1,
        'label' : 1,
        'result_type': 'true_positive'
    },
]
```

```python
result_types = pandas.DataFrame(result_types).set_index(['prediction', 'label'])
result_types
```

```python
data_test = data_test.join(result_types, on = ['prediction', 'label'], how = 'left')
data_test
```

```python
data_test.to_csv('./results/logistic_regression_predictions.test.csv')
```

```python
report = sklearn.metrics.classification_report(
    y_true=data_test['label'],
    y_pred=data_test['prediction'],
    output_dict=True  # Get the report in dictionary format
)

# Convert the dictionary to a pandas DataFrame
report_df = pandas.DataFrame(report).transpose()
print(report_df)
```

```python
confusion_matics_from_sklearn = sklearn.metrics.confusion_matrix(
    y_true = data_test['label'],
    y_pred = data_test['prediction'],
)
confusion_matics_from_sklearn
```

```python
scores = {
    "true_positive": 1845,
    "true_negative": 2391,
    "false_negative": 347,
    "false_positive": 642,
    "accuracy": 0.810718,
    "precision": 0.741858,
    "recall": 0.841697,
    "support": 2192,
    "f1-score": 0.788630
}
```
