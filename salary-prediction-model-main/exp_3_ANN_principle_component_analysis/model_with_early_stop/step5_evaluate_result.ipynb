```python
import numpy
import pandas
import joblib
import json
import seaborn

import sklearn

%matplotlib inline
```

```python
data_test = pandas.read_csv('../data/features.test.csv').set_index('id')
data_test
```

```python
model = joblib.load('../model/model_with_early_stop.joblib')
model
```

```python
data_test['prediction']= model.predict(data_test.drop(['label'], axis= 'columns'))
data_test
```

```python
data_test.to_csv('../results/model_with_early_stop_predictions.test.csv')
```

```python
confusion_matics_from_sklearn = sklearn.metrics.confusion_matrix(
    y_true = data_test['label'],
    y_pred = data_test['prediction'],
)
confusion_matics_from_sklearn
```

```python
report = sklearn.metrics.classification_report(
    y_true=data_test['label'],
    y_pred=data_test['prediction'],
    output_dict=True  # Get the report in dictionary format
)

# Convert the dictionary to a pandas DataFrame
report_df = pandas.DataFrame(report).transpose()
print(report_df)
```

```python
scores = {
     'true_positive': 1733,
     'true_negative': 2533,
     'false_negative': 459,
     'false_positive': 500,
     'accuracy': 0.816459,
     'precision':  0.776086,
     'recall': 0.790602,
     'support': 2192,
     'f1-score': 0.783277
}
```
