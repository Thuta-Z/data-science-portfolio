```python
import pandas
import sklearn.model_selection
```

```python
data_train_test = pandas.read_csv('./data/high_salary.csv').set_index('id')
data_train_test
```

```python
data_train, data_test = sklearn.model_selection.train_test_split(
    data_train_test, 
    train_size   = 0.75, 
    shuffle      = True, 
    random_state = 0
)
data_train.to_csv('./data/high_salary.train.csv')
data_test.to_csv('./data/high_salary.test.csv')
```
