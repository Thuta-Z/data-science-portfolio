```python
import joblib
import json
import pandas

import sklearn.neural_network
```

```python
data_train= pandas.read_csv('../data/features.train.csv').set_index('id')
data_train
```

```python
model = sklearn.neural_network.MLPClassifier(
    hidden_layer_sizes    = (20, 10),
    solver                = 'sgd', 
    activation            = 'logistic',
    alpha                 = 0.0,
    batch_size            = 32,
    learning_rate         = 'constant',
    learning_rate_init    = 0.1,
    max_iter              = 1000,
    momentum              = 0.0, 
    nesterovs_momentum    = False, 
    validation_fraction   = 0.0, 
    shuffle               = True,
    random_state          = 0
)
model
```

```python
model.fit(
    data_train.drop(['label'], axis = 'columns'),
    data_train['label']
)
```

```python
joblib.dump(model,'../model/default_model.joblib' )
```

```python
model.get_params()
```

```python
with open('../model/default_model_config.json', 'w') as f:
    json.dump(model.get_params(), f)
```
